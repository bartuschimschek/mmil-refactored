# Load the dataset

```python
scrna = sc.read_h5ad('../data/10xpbmc10k-2020/expressions.h5ad')
scrna
```

```python
scatac = sc.read_h5ad('../data/10xpbmc10k-2020/peaks.h5ad')
scatac
```

# Configure and train the model

```python
model = scmulti.models.MultiVAE(
    adatas=[[scrna], [scatac]],
    names=[['scRNA-seq'], ['scATAC-seq']],
    pair_groups=[[0], [0]],
    z_dim=20,
    h_dim=128,
    hiddens=[],
    shared_hiddens=[],
    adver_hiddens=[],
    recon_coef=1,
    kl_coef=1e-5,
    integ_coef=1e-2,
    cycle_coef=0,
    adversarial=False,
    dropout=0.2,
)
```

```python
model.train(
    n_iters=10000,
    batch_size=64,
    lr=3e-4,
    val_split=0.1,
    adv_iters=0,
    validate_every=1300
)
```

# Validation

```python
z = model.predict(
    adatas=[[scrna], [scatac]],
    names=[['scRNA-seq'], ['scATAC-seq']],
    batch_size=64,
)
z
```

```python
sc.pp.neighbors(z)
sc.tl.umap(z)
```

```python
sc.pl.umap(z, color=['modality', 'cell_type'], ncols=1)
```
